```python
#use template from TwythonStreamer class, redefine what to do on success
class MyTwythonStreamer(TwythonStreamer):
    def on_success(self, data):
        if data['retweeted'] == 'True':
            return
        if data['lang'] == 'en':
            tweets.append(data)
            print('received tweet #{}'.format(len(tweets)))
        if len(tweets) >=20:
              self.disconnect()
    def on_error(self, status_code, data):
        print(status_code, data)
        self.disconnect()
```
